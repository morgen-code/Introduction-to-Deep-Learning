```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ステップ関数
def step_func(x):
  return np.where(x >= 0, 1, 0)

x = np.linspace(-4, 4, 500)

plt.plot(x, step_func(x))
plt.legend(["step"], loc = "best")
plt.yticks(np.arange(0, 1.2, step = 0.5))
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# シグモイド関数
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

x1 = np.linspace(-10, 10, 500)

plt.plot(x1, sigmoid(x1))
plt.legend(["sigmoid"], loc = "best")
plt.yticks(np.arange(0, 2, step = 0.5))
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# 双曲線正接関数
def tanh(x):
  return np.tanh(x)

x1 = np.linspace(-10, 10, 500)

plt.plot(x1, tanh(x1))
plt.legend(["tanh"], loc = "best")
plt.yticks(np.arange(-1.5, 2, step = 0.5))
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# ReLU関数
def relu_func(x):
  return np.maximum(0, x)

x1 = np.linspace(-6, 6, 500)

plt.plot(x1, relu_func(x1))
plt.legend(["ReLU"], loc = "best")
plt.yticks(np.arange(0, 6.5, step = 0.5))
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# まとめて表示
x1 = np.linspace(-4, 4, 500)

plt.plot(x1, step_func(x1))
plt.plot(x1, sigmoid(x1))
plt.plot(x1, tanh(x1))
plt.plot(x1, relu_func(x1))
plt.legend(["step", "sigmoid", "tanh", "ReLU"], loc = "best")
plt.yticks(np.arange(-1, 5, step = 0.5))
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
